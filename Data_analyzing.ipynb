```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DF = pd.read_csv("2020-2025.csv")
df = DF.copy()
df
```

```python
df.isnull().sum()
```

```python
# getting the top 10 countries with the highest gdp in the year 2020
Df = df.sort_values("2020", ascending=False)
Df.reset_index(drop=True, inplace=True)
Df.loc[0:10, "2020"]
```

```python
countries = Df.loc[0:9, "Country"].to_list()
values = Df.loc[0:9, "2020"].to_list()
values
```

```python
# Comparision of Countries with the highest gdp in the year 2020
countries = Df.loc[0:9, "Country"].to_list()
values = Df.loc[0:9, "2020"].to_list()

plt.figure(figsize=(13,6))
plt.bar(countries, values)
for i in range(len(countries)):
    plt.text(i, values[i] + 0.005*max(values), str(values[i]),fontsize=12,fontweight="bold", ha="center", va= "bottom")

plt.xlabel("Top 10 countries with the highest GDP in the year 2020")
plt.ylabel("GDP Values")
plt.title("Comparision of Countries with the highest gdp in the year 2020")
plt.tight_layout()
plt.style.use('seaborn-v0_8-darkgrid')
plt.show()
```

```python
# Comparision of Countries with the highest gdp in the year 2025
countries = Df.loc[0:9, "Country"].to_list()
values = Df.loc[0:9, "2025"].to_list()

plt.figure(figsize=(13,6))
plt.bar(countries, values, color="#0AF50A")
for i in range(len(countries)):
    plt.text(i, values[i] + 0.015*max(values), str(values[i]), fontsize = 10, fontweight="bold", ha="center", va="center")

plt.xlabel("Top 10 countries with the highest GDP in the year 2025")
plt.ylabel("GDP Values")
plt.title("Comparision of Countries with the highest gdp in the year 2025")
plt.show()
```

```python
# GDP Growth of the top 10 countries with the highest gdp in the last 5 years
years = ["2020","2021","2022","2023","2024","2025"]
countries = Df.loc[0:9, "Country"].to_list()
plt.figure(figsize=(10,6))

for _,rows in Df.loc[0:9].iterrows():
    gdp_values = [rows[year] for year in years]
    plt.plot(years, gdp_values, marker="o", label=rows["Country"])

plt.xlabel("Years")
plt.ylabel("GDP Values")
plt.title("GDP growth of the top 10 countries with the highest gdp in the last 5 years")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```
